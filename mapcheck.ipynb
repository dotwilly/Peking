```python
import requests
import json
from baidumap.api.handle import get_handle
```

```python
ak_key = 'tDfnKwkckv1yqTnrpBwkSkPvVO2DV2xa'
place_search = get_handle(ak_key, 'place/v2/search', is_list=True)
place_search.set_params(query='星巴克', region='北京市')
atm_in_beijing = place_search.run(max_page_num=-1, page_size=10, max_result_num=-1, interval=0.5)
print(atm_in_beijing)
print(atm_in_beijing.get_property('address'))
```

```python
lat=atm_in_beijing.get_property('lat')
print(lat)
lng=atm_in_beijing.get_property('lng')
print(lng)
name=atm_in_beijing.get_property('name')
print(name)
area=atm_in_beijing.get_property('area')
print(area)
```

```python
import jsonpath
lat = jsonpath.jsonpath(lat,'$..lat')   #不管有多少层，写两个.都能取到
print(lat) #['xiaohei'] 返回的是一个列表
lng=jsonpath.jsonpath(lng,'$..lng')
print(lng)
name=jsonpath.jsonpath(name,'$..name')
print(name)
area=jsonpath.jsonpath(area,'$..area')
print(area)
```

```python
print(len(lat))
print(len(lng))
print(len(name))
print(len(area))
```

```python
plot_num_estimators= []
plot_result_accuracy = []
realname=[]
realarea=[]
for i in range(0,len(lat)):
    plot_result_accuracy.append(lat[i])
    plot_num_estimators.append(lng[i])
    realname.append(name[i])
    realarea.append(area[i])
```

```python
import pandas as pd
team1 = pd.DataFrame(plot_result_accuracy)
team2 = pd.DataFrame( plot_num_estimators)
team3 = pd.DataFrame( realname)
team4 = pd.DataFrame( realarea)
team1.columns = ["lat"]
team2.columns = ["lng"]
team3.columns=["name"]
team4.columns=["area"]

teams = pd.concat([team3,team1,team2], axis=1)
teams_area = pd.concat([team4,team3,team1,team2], axis=1)
```

```python
print(teams.shape)
print(teams.head(5))
print(teams_area.head(5))
print(teams_area.shape)
```

```python
teams.to_csv('./starbucks_teams.csv')
teams_area.to_csv('./starbucks_area.csv')
```
